```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip()) #strip is for removing spaces / \n / \t types things
```

```python
#to extract tags likes 'p'...there are many lines in webpages which have tag p...so differentiation is difficult to extract exactly which p tag do we want...

#in order to remove this problem we use CLASS in html to differentiate
```

```python
soup.find_all('p')
```

```python
for i in soup.find_all('div',class_="rating_star_container"):
    print(i.text.strip())
```

```python
for i in soup.find_all('span',class_="companyCardWrapper__companyRatingCount"):
    print(i.text.strip())
```

```python
for i in soup.find_all('div',class_="companyCardWrapper__primaryInformation"):
    for j in i:
        print(j.text.strip())
        print("\n")
```

```python
len(soup.find_all('div',class_="companyCardWrapper__primaryInformation"))
```

```python
company = soup.find_all('div',class_="companyCardWrapper__primaryInformation")
```

```python
name=[]
rating=[]
review=[]
values=[]
for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div',class_="rating_star_container").text.strip())
    review.append(i.find('span',class_="companyCardWrapper__companyRatingCount").text.strip())
    values.append(i.find('span',class_="companyCardWrapper__ratingValues").text.strip())
# print(name)
# print(rating)
# print(review)
# print(values)
d = {'name':name, 'rating':rating,'review':review, 'values':values}
df = pd.DataFrame(d)
```
